```python
df_sort = df.sort_values(by=['employee_name', 'email'], na_position='last')
df_sort['email'] = df_sort['email'].fillna('placeholder')
df_sort['employee_name'] = df['employee_name'].str.strip()
df_sort['email'] = df['email'].str.strip()
df_sort['email'] = df['email'].str.rstrip("\n")
df_sort['email'] = df['email'].str.lstrip("\n")
df_sort = df[df['email'].str.contains('@', na=False) | df['email'].isna()]
```

```python
df_unique = df_sort.drop_duplicates(subset=['employee_name', 'email'], keep='last')
df_unique.loc[df_unique['email'] == 'placeholder', 'email'] = pd.NA
```

```python


# Select the missing columns
missing_columns = df_unique.loc[1:DFrows, ['employee_id','email', 'telephone']]

# Select the desired range of rows and columns
employeeInformation = df_unique.loc[1:DFrows, 'employee_name':'languages']

# Concatenate the DataFrames along the columns
employeeInformation = pd.concat([employeeInformation, missing_columns], axis=1)
employeeInformation.tail(5)
```

```python
employeeEducation = df_unique.loc[1:DFrows,'education_date_1':'professional_training']
missing_columns = df_unique.loc[1:DFrows, ['employee_id']]
employeeEducation = pd.concat([employeeEducation, missing_columns], axis=1)
employeeEducation.head()
```

```python
# Select the desired range of rows and columns
employeeExperience = df_unique.loc[1:DFrows, 'name_current_employer':'duties_20']
missing_columns = df_unique.loc[1:DFrows, ['employee_id']]

# Drop the 'email' and 'telephone' columns if they exist in the selection
employeeExperience = employeeExperience.drop(columns=['email', 'telephone'], errors='ignore')
employeeExperience = pd.concat([employeeExperience, missing_columns], axis=1)

employeeExperience.head()
```

```python
TransactionTimestamps = df_unique.loc[1:DFrows,['employee_name','Date Created','Date Updated']]
missing_columns = df_unique.loc[1:DFrows, ['employee_id']]
TransactionTimestamps = pd.concat([TransactionTimestamps, missing_columns], axis=1)
TransactionTimestamps.head(4)


```
